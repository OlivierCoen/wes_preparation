```python
import pandas as pd
from tqdm.auto import tqdm
```

```python
gene_file = "/Phytozome/PhytozomeV12/Ahypochondriacus/annotation/Ahypochondriacus_315_v1.0.gene.gff3"
repeat_file = "Phytozome/PhytozomeV12/Ahypochondriacus/annotation/Ahypochondriacus_315_v1.0.repeatmasked_assembly_v1.0.gff3"
```

```python
GTF_COLS = ["seqid", "source", "type", "start", "end", "score", "strand", "phase", "attributes"]
```

```python
gene_df = pd.read_csv(gene_file, sep="\t", header=None, names=GTF_COLS, skiprows=3)
repeat_df = pd.read_csv(repeat_file, sep="\t", header=None, names=GTF_COLS, skiprows=1)
```

```python
cds_df = gene_df[gene_df.type == "CDS"]
```

```python
def check_overlap(sub_gene_df, overlaps, contig, start, end):
    match_sub_gene_df = sub_gene_df[(sub_gene_df.start >= start) & (sub_gene_df.end <= end)]
    if len(match_sub_gene_df) > 0:
        match_sub_gene_df['contig'] = contig
        overlaps += match_sub_gene_df.to_dict(orient="records")
        print(len(overlaps))
```

```python
overlaps = []
for contig in tqdm(cds_df.seqid.unique()):
    sub_gene_df = cds_df[cds_df.seqid == contig][["start", "end"]]
    sub_repeat_df = repeat_df[repeat_df.seqid == contig][["start", "end"]]
    sub_repeat_df.apply(lambda row: check_overlap(sub_gene_df, overlaps, contig, row.start, row.end), axis=1)
    
```
