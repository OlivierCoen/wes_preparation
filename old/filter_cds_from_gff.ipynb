```python
import pandas as pd
from tqdm.auto import tqdm
```

```python
GTF_COLS =  [
    "seqname", 
    "source", 
    "feature", 
    "start", 
    "end", 
    "score", 
    "strand", 
    "frame", 
    "attribute"
]
```

```python
annotation_file = "Ahypochondriacus_459_v2.1.gene.gff3"
```

```python
df = pd.read_csv(annotation_file, sep="\t", header=None, names=GTF_COLS, skiprows=3)
df
```

```python
df = df[df["feature"] == "mRNA"]
df["length"] = df["end"] - df["start"] + 1
print(df["length"].sum())
df["gene"] = df["attribute"].str.split(";").str[1].str.split("Parent=").str[1]
df["longest"] = df["attribute"].str.split(";").str[3].str.split("longest=").str[1]
df[df["longest"] == 0]
```

```python
df.sort_values(by=["gene", "start", "end"]).groupby("gene").agg(list)
```
